# Experiment: KS Test
The following code block reads in the properties of stars from the referenced sources, calculates values of q by dividing the secondary mass from the primary mass, counts these values, then accounts for completeness by constraining the distriubtion to the value that the reference materials are complete to. These constrained distributions are then counted, and the 2 sample KS test is performed to compare to see if the two distributions come from the same parent distribution. 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ks_2samp 

# importing data
s_DI, e_s_DI, pm_DI, sm_DI = np.loadtxt('cmrd_data_DI.txt', skiprows = 2, delimiter = ',', unpack = True) # separation (AU), error, primary mass (Solar), secondary mass (Solar)
q_DI = np.zeros(124)
q_DI = np.divide(sm_DI, pm_DI) # complete for q > .2
pm_RV, sm_RV, s_RV, e_s_RV = np.loadtxt('cmrd_data_RV.txt', skiprows = 2, delimiter = ' ', unpack = True) # primary mass (Solar), secondary mass (Solar), separation (AU), error 
q_RV = np.zeros(214)
q_RV = np.divide(sm_RV, pm_RV, q_RV) # complete for q > .4
p_DI, e_p_DI, m_DI = np.loadtxt('mag_sep_DI.txt', skiprows = 0, delimiter = ',', unpack = True) # separation ("), error, delta magnitude of DI
period_RV = np.loadtxt("period_RV.txt", skiprows = 0, unpack = True)


# counting data (each array for survey is same size)
count_DI = s_DI.shape # = 124
count_RV = s_RV.shape # = 214

# finding index of values of q beyond completeness limit
sorted_q_DI = np.sort(q_DI)
sorted_q_RV = np.sort(q_RV)
for pos, val in enumerate(sorted_q_DI):
    if val > .4:
        complete_pos_DI = pos # = 31
#         print(complete_pos_DI)
        break
for pos, val in enumerate(sorted_q_RV):
    if val > .4:
        complete_pos_RV = pos # = 20
#         print(complete_pos_RV)
        break

# places associated q and s values together in one 2 row array
DI_values = np.stack((q_DI, s_DI, p_DI, m_DI)) # 4 rows, 124 values each
RV_values = np.stack((q_RV, s_RV, sm_RV, period_RV))

# sorts array of values by ascending q, keeps associated values together
a1 = DI_values[0,:].argsort() 
DI_values_sorted = DI_values[:,a1]
a2 = RV_values[0,:].argsort() 
RV_values_sorted = RV_values[:,a2]

# splicing off values beyond completeness range using index found above, therefore accounting for completeness
DI_values_complete = np.array([DI_values_sorted[i][31:] for i in range(0,4)])
RV_values_complete = np.array([RV_values_sorted[i][20:] for i in range(0,4)])

DI_values_complete_count = DI_values_complete[0].shape # = 93
RV_values_complete_count = RV_values_complete[0].shape # = 194

# ks test
# print(ks_2samp(s_q_DI_complete[0], s_q_RV_complete[0])) # statistic=0.10576923076923077, pvalue=0.61797345550906002

# plotting Q vs. S
plt.scatter(DI_values_complete[1], DI_values_complete[0])
plt.xscale('log')
plt.title("DI - Q vs. S (complete)")
plt.xlabel("Separation (AU)")
plt.ylabel("Q (secondary mass / primary mass)")
plt.show()

plt.scatter(RV_values_complete[1], RV_values_complete[0])
plt.xscale('log')
plt.title("RV - Q vs. S (complete)")
plt.xlabel("Separation (AU)")
plt.ylabel("Q (secondary mass / primary mass)")
plt.show()

# frequency constrained to completeness limit
frequency_DI = 93 / 449 # = .2071
frequency_RV = 194 / 761 # = .2549

# plotting delta mag vs sep (") for DI
plt.scatter(DI_values_complete[3], DI_values_complete[2])
plt.title("DI - Delta Mag vs. Sep (\") (complete)")
plt.yscale('log')
plt.xlabel("Delta Magnitude")
plt.ylabel("Separation (\")")
plt.show()

# plotting secondary mass vs. period for RV
plt.scatter(RV_values_complete[2], RV_values_complete[3])
plt.title("RV - Secondary Mass vs. Period")
plt.yscale('log')
plt.xlabel("Secondary Mass (Solar Masses)")
plt.ylabel("Period (years)")
plt.show()
```

# Conclusion: KS Test
The results of the KS test tell us that we are unable to reject the null hypothesis that the samples came from the same distribution. So, we assume they did come from the same distribution. In the context of this experiment, this means that the mass ratio distribution does not depend on orbital separation.
